### 감정세기가 뭘 의미하는지 모르겠지만 중요한 것 같지 않아 삭제함

```python
realdata = data.loc[:,['발화문','상황']]
realdata.head()
```

```python
data['상황'].unique()
```

```python
realdata.loc[(realdata['상황'] == "anger"), '상황'] = 0  #분노 => 0
realdata.loc[(realdata['상황'] == "sad"), '상황'] = 1  #슬픔 => 1
realdata.loc[(realdata['상황'] == "fear"), '상황'] = 2  #공포 => 2
realdata.loc[(realdata['상황'] == "disgust"), '상황'] = 3  #역겨움 => 3
realdata.loc[(realdata['상황'] == "neutral"), '상황'] = 4  #중립 => 4
```

```python
realdata.head()
```

```python
data_list = []
for q, label in zip(realdata['발화문'], realdata['상황']):
    data = []
    data.append(q)
    data.append(str(label))

    data_list.append(data)
```

```python
data_list
```

```python
from sklearn.model_selection import train_test_split
dataset_train, dataset_test = train_test_split(data_list, test_size=0.25, random_state=0)

print(len(dataset_train)) #10954 출력
print(len(dataset_test)) #3652 출력
```

## 데이터 전처리(토큰화, 정수 인코딩, 패딩)

- 토큰화 : 주어진 말뭉치에서 토큰(token)이라 불리는 단위로 나누는 작업이다. 토큰의 단위는 상황에 따라서 다르게 적용될 수 있다.

- 정수 인코딩 : 텍스트에 등장하는 단어의 빈도수에 따라 정렬하여, 자주 등장하는 단어에 정수를 부여하여 컴퓨터가 처리하기 쉽게 만들어 주는 것이다.
